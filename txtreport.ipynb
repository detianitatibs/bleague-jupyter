```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
URL_TOP = 'https://sportsnavi.ht.kyodo-d.jp/basketball/stats/b1/score/4176/'
URL_TXT =  URL_TOP+'txt/'
```

```python
html = urllib.request.urlopen(URL_TXT)
top = BeautifulSoup(html, 'html5lib')
```

```python
liveBody = top.find(class_="liveBody")
```

```python
type(liveBody)
```

```python
# 各クォーターの名称を取得
quoter = liveBody.find_all(class_="ht-modAccordion__title")
```

```python
# 各クォーター毎の要素を取得 
gameterm = liveBody.find_all(class_="ht-gameTerm")
```

```python
# 各要素からリストを作成する
timeline = []
for i in range(len(quoter)):
    tgt_quoter = quoter[i].string
    tgt_linetime = gameterm[i].find_all(class_="ht-modTimeline__time")
    tgt_linename = gameterm[i].find_all(class_="ht-modTimeline__name")
    tgt_linescore = gameterm[i].find_all(class_="ht-modTimeline__score")
    tgt_linedesc = gameterm[i].find_all(class_="ht-modTimeline__desc")
    for j in range(len(tgt_linetime)):
        timeline.append([
            tgt_quoter,
            # 改行を削除
            tgt_linetime[j].string.replace('\n',''),
            # 2つ以上空白なら削除する
            tgt_linename[j].string.replace('\n','').replace('  ',''),
            tgt_linescore[j].string.replace('\n','').replace('  ',''),
            tgt_linedesc[j].string.replace('\n','').replace('  ','')])
```

```python
# 文字列でリスト取得まで完了
timeline
```
